```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import os

# File to Load (Remember to Change These)
file_to_load = "/Users/mignonduplessis/Documents/BootCampRep/Pandas_Challenge/HeroesOfPymoli/Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data.head()
```

## Player Count

* Display the total number of players


```python
purchase_data_df = purchase_data.groupby("SN")["SN"].unique()
total_players = purchase_data_df.count() 

purchase_data_df = pd.DataFrame([{ "Number of Players": total_players}])
purchase_data_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
from pandas import DataFrame

unique_items = purchase_data['Item ID'].nunique()
average = purchase_data['Price'].mean()
purchase_count = purchase_data['Price'].count()
total_rev = purchase_data['Price'].sum()



d = {'Number of Unique Items':[unique_items], 'Average Price':[average] , 'Number of Purchases':[purchase_count], 'Total Revenue':[total_rev]}


df = DataFrame(d, columns= ['Number of Unique Items','Average Price','Number of Purchases','Total Revenue'])
round_df = df.round(decimals=2)


print(round_df)

round_df.style.hide_index()

```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
from pandas import DataFrame


no_duplicate_players = purchase_data.drop_duplicates(['SN'], keep ='last')
grouped_gender_df = no_duplicate_players.groupby(['Gender'])


## Filter the data so that only those sightings in Male are in a DataFrame
male_df = grouped_gender_df.loc[grouped_gender_df["Gender"] == "Male", :]
gender_count = male_df['Gender'].value_counts()
total = male_df['Gender'].count()
percentage = (gender_count/total)*100
rounding=percentage.round(2)

#
female_df = grouped_gender_df.loc[grouped_gender_df["Gender"] == "Female", :]
female_count = female_df['Gender'].value_counts()
f_total = female_df['Gender'].count()
f_percentage = (female_count/f_total)*100
f_rounding = f_percentage.round(2)

#
otehr_df = grouped_gender_df.loc[grouped_gender_df["Gender"] == "Other / Non-Disclosed", :]
o_gender_count = otehr_df['Gender'].value_counts()
o_total = male_df['Gender'].count()
o_percentage = (o_gender_count/o_total)*100
o_rounding=o_percentage.round(2)

#
df_gender = {"Total Count": [gender_count, female_count , o_gender_count],
                "Percentage of Players": [rounding, f_rounding, o_rounding]}

gender_summary_table = DataFrame(df_gender, columns= ['Total Count',"Percentage of Players"])
grouped_gender_df = gender_summary_table.groupby(['Gender'])
print(grouped_gender_df)
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create the bins in which Data will be held
bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
# Create the names for the bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
# Place the data series into a new column inside of the DataFrame
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=group_names)
purchase_data.head()
```

```python
age_totals = purchase_data["Age Group"].value_counts()
percentages = age_totals / total_players * 100
age_demographics = pd.DataFrame({"Total Count": age_totals, "Percentage of Players": percentages})

# Presenting the data
age_demographics = age_demographics.round(2)

# Display Age Demographics Table
age_demographics.sort_index()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#find total unique players per bin
total_players_in_bin=purchase_data.groupby('Age Group')["SN"].nunique()

#apply functions to new dataframe column values
total_pv=purchase_data.groupby('Age Group')["Price"].sum()
avg_total_pv =purchase_data.groupby('Age Group')["Price"].mean()
purchase_pp = total_pv/total_players_in_bin

#set dataframes
age_demographics = pd.DataFrame({"Total Count": age_totals})
purchasing_analysis= pd.DataFrame ({"Average Purchase Price":avg_total_pv,"Total Purchase Value":total_pv, "Avg Total Purchase per Person":purchase_pp })

#round dataframes to two decimals
age_demographics=age_demographics.round(2)
purchasing_analysis=purchasing_analysis.round(2)

#merge dataframes
pd.concat([age_demographics,purchasing_analysis], axis=1, ignore_index=False, sort=False)
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#apply functions to new dataframe column values
purchase_count=purchase_data.groupby(["SN"]).nunique()["Price"]
average_purcahse_price=purchase_data.groupby(["SN"]).mean()["Price"]
total_purchase_value=purchase_data.groupby(["SN"]).sum()["Price"]

#set dataframe
top_spenders_df = pd.DataFrame({"Purchase Count":purchase_count,"Average Purchase Price":average_purcahse_price,"Total Purchase Value":total_purchase_value})

#set decending order in dataframe column "Total Purchase Value"
top_spenders_df = top_spenders_df.sort_values("Total Purchase Value",ascending=False)

#round dataframes to two decimals
top_spenders_df = top_spenders_df.round(2)

top_spenders_df.head()
```
